We're filling the missing values in the "average" column by calculating the batting average for each player. Since batting average is calculated by dividing the total runs scored by the number of times dismissed (outs), we're using this formula:total_runs/out to impute the missing values. This approach ensures that each player's average reflects their performance based on the runs scored and the number of times they were dismissed.

```python
# Filling the missing values in average by dividing total_runs by out
average_values = df3['total_runs'] / df3['out']

# Replacing the infinite values with 'can't define' and NaNs with 0
average_values = average_values.replace([np.inf, -np.inf], "can't define").fillna(0)

# Updating the average column with the calculated values
df3['average'] = df3['average'].fillna(average_values)
```

# Correlation Analysis

Correlation analysis examines the relationship between pairs of variables in a dataset, determining the extent to which changes in one variable correspond to changes in another, measured by the correlation coefficient, which ranges from -1 to 1.


```python
# Drop the 'match_id' column from df1
df1_without_id = df1.drop(columns=['match_id'])

# Select only numeric columns for correlation
numeric_columns_df1 = df1_without_id.select_dtypes(include=['int64', 'float64']).columns

# Calculate correlation matrix
corr_matrix_df1 = df1_without_id[numeric_columns_df1].corr()

# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix_df1, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix for df1 (Excluding 'match_id')")
plt.show()
```

```python
# Drop the 'id' column from df2
df2_without_id = df2.drop(columns=['id'])

# Select only numeric columns for correlation
numeric_columns_df2 = df2_without_id.select_dtypes(include=['int64', 'float64']).columns

# Calculate correlation matrix
corr_matrix_df2 = df2_without_id[numeric_columns_df2].corr()

# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix_df2, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix for df2 (Excluding 'id')")
plt.show()

```

# DATA AGGREGATION

Data aggregation involves the process of gathering and summarizing data from multiple sources or across various dimensions to provide meaningful insights or simplify complex datasets.

Here, we're combining the df1,df2,df3 to one dataset df

```python
df = pd.merge(df1, df2, left_on='match_id', right_on='id', how='left')
```

```python
df = pd.merge(df, df3, on='batsman', how='left')
```

```python
df.info()
```

```python
#data aggregation that calculates the total runs scored by each batting team:
# Grouping the data by 'batting_team' and calculating the total runs scored
batting_team_total_runs = df.groupby('batting_team')['total_runs_currentdelivery'].sum()

# Displaying the aggregated data
print("Total Runs Scored by Each Batting Team:")
print(batting_team_total_runs)
```

```python
#data aggregation that calculates the average strike rate of each batsman:
# Grouping the data by 'batsman' and calculating the average strike rate
batsman_average_strike_rate = df.groupby('batsman')['strikerate'].mean()

# Displaying the aggregated data
print("Average Strike Rate of Each Batsman:")
print(batsman_average_strike_rate)
```

# **Bar chart**
A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.


# When to use it ?
Compare categorical data.  
Comparisons among discrete categories.  
One axis of the chart shows the specific categories being compared, and the other axis represents a measured value.  

# Question 1: Which team has won the most matches in the dataset?


```python
# Define custom colors for each team
custom_colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'cyan', 'magenta']

win_counts = df['winner'].value_counts().head(10)
win_counts.plot(kind='bar', figsize=(12, 6), color=custom_colors) 
plt.title('Top 10 Winning Teams')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45, ha='right')  # Rotate the x-axis labels further and align them to the right
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

# Marks and Channels Used:
 

Marks:  
Vertical bars represent the number of wins for each team.  

Channels:  
Position (X-axis): Represents the teams.  
Position (Y-axis): Represents the number of wins.  
Length (Bars): Indicates the magnitude of wins for each team.  
Color (Bars): Could differentiate between different teams, although in this specific plot, all bars are the same color.

# Observations

The bar plot illustrates the number of wins for the top 10 winning teams in the dataset. Each bar represents a team, and its length corresponds to the number of wins accumulated by that team. From the visualization, it is evident that Mumbai Indians hold the highest number of wins, followed by Chennai Super Kings and Kolkata Knight Riders. The plot provides a clear comparison of the success rates among the top-performing teams, offering insights into their dominance in the league. Overall, it showcases the competitive landscape of the teams in the dataset, with Mumbai Indians emerging as the most successful team based on the number of wins.

# Question 2: How does the average run rate vary across different seasons?


```python
df['run_rate'] = df['total_runs_currentdelivery'] / df['numberofballs']  # Calculate run rate
season_avg_run_rate = df.groupby('Season')['run_rate'].mean()
season_avg_run_rate.plot(kind='line', figsize=(10, 6), marker='o', color='green')
plt.title('Average Run Rate Across Seasons')
plt.xlabel('Season')
plt.ylabel('Average Run Rate')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

Channels:   
Position (X-axis): Represents the seasons.  
Position (Y-axis): Represents the average run rate for each season.  
Color (Line): The color of the line represents the data series (average run rate).

# Observations:

The line plot depicts the average run rate across different seasons of the IPL (Indian Premier League). It illustrates the trend of run rates over time, providing insights into the scoring patterns in each season. From the graph, it's evident that there are fluctuations in the average run rate from season to season, with some seasons witnessing higher run rates compared to others. # Stacked area chart

# Question 3: What is the trend of total runs scored by each team over different seasons?


```python
team_season_runs = df.groupby(['Season', 'batting_team'])['total_runs_currentdelivery'].sum().unstack(fill_value=0)
team_season_runs.plot(kind='area', stacked=True, figsize=(12, 6))
plt.title('Total Runs Scored by Each Team Over Seasons')
plt.xlabel('Season')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.legend(title='Team', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# Marks and Channels Used:

Marks:  
Filled areas represent the total runs scored by each team.   

Channels:  
Position (X-axis): Represents the seasons.  
Position (Y-axis): Represents the total runs scored.  
Color (Areas): Differentiates between teams. Each team's area is filled with a different color.  

# Observations:

The stacked area chart displays the total runs scored by each team over different seasons in the Indian Premier League (IPL). Each color represents a different team, and the height of each color segment at any given point on the x-axis represents the total runs scored by that team in a particular season.

From the chart, we can observe the overall trend of runs scored by each team across seasons. For instance, teams like Mumbai Indians, Chennai Super Kings, and Royal Challengers Bangalore consistently appear to score higher totals across multiple seasons, indicating their strong batting performances. On the other hand, some teams like Kochi Tuskers Kerala and Pune Warriors show lower total runs, possibly due to their shorter presence in the IPL or lower overall performance.

Additionally, fluctuations in the height of the colored segments within each season reveal variations in the performance of teams from one season to another. This visualization provides insights into the batting prowess of each team over time and can help in analyzing the historical performance and trends of IPL teams in terms of run-scoring abilities.

# Question 4: How does the frequency of matches played vary across different venues?

```python
venue_counts = df['venue'].value_counts().head(10)  # Get the top 10 venues by match frequency

venue_df = df[df['venue'].isin(venue_counts.index)]
venue_counts_stacked = venue_df.groupby(['venue', 'winner']).size().unstack(fill_value=0)

venue_counts_stacked.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Match Outcomes by Venue (Top 10 Venues)')
plt.xlabel('Venue')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.legend(title='Winner', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# Marks and Channels Used:

Marks:  
Rectangular bars represent each category (venue).  
Different segments within each bar represent different categories (winners).  

Channels:  
Position (X-axis): Represents the venues where matches were played.  
Position (Y-axis): Represents the number of matches played.  
Color (Segments): Differentiates between winners. Each segment of the bar is colored according to the winning team.

# Observations:


The stacked bar chart representation above shows the frequency of matches played at the top 10 venues, along with the number of matches won by each team at those venues. Each row corresponds to a specific venue, and each column represents a team that has won matches at that venue. For example, at the Eden Gardens venue, Kolkata Knight Riders won the highest number of matches, followed by Chennai Super Kings,and others. Similarly, the bar chart provides insights into the match outcomes at other popular venues such as Feroz Shah Kotla, M Chinnaswamy Stadium, and Wankhede Stadium. This table offers a comprehensive view of the match frequency and team performances across different venues, aiding in the analysis of venue preferences and team strengths in various locations.
